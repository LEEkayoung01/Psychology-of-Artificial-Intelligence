```python
# 3. Generate one million random single digit numbers and find out how many sequences of numbers that matches exactly the last 6 digits of your student number. 
import numpy as np
arr1 = []
arr2 = []
time = 1000
for time in range(time):
    number = np.random.randint(low = 0, high = 10, size = 1000000)
    target = [1, 3, 0, 4, 3, 1]
    count = 0
    j = 0
    while j < len(number):
        i = 0
        while number[j] == target[i]:
            arr1.append(number[j])
            if i == 5:
                count+=1
                break
            if j == len(number)-1:
                break
            i+=1
            j+=1

        if number[j] != target[i]:
            j+=1
        if j == len(number)-1:
            break

    arr2.append(count)

# print(arr1)
print(arr2)

probability=(time-arr2.count(0))/time
print(probability)
```

```python
# 4. housing.csv
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv")
#data.head()

# Median house value, Median income 2 
plt.scatter(data['median_house_value'], data['median_income'], s=3)
plt.xlabel('Median House Value')
plt.ylabel('Median Income')
plt.show()

# longgitude, latitude, median_house_value 3 
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['longitude'], data['latitude'], data['median_house_value'], s=3)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Median House Value')
```

```python
# 4-(a) Pearson correlation coefficient
# Dividing the dataset
data_1 = data.sample(frac=0.5)
data_2 = data.drop(data_1.index)

# Calculating Pearson's correlation coefficient
r1 = np.corrcoef(data_1['median_house_value'], data_1['median_income'])
print("Pearson correlation coefficient of data_1: "+str(r1[1,0]))
r2 = np.corrcoef(data_2['median_house_value'], data_2['median_income'])
print("Pearson correlation coefficient of data_2: "+str(r2[1,0]))
```

```python
# 4-(b)
from matplotlib import pyplot as plt
data_total_bedrooms = data['total_bedrooms']
df=pd.DataFrame()
categories=[0, 200, 400, 600, 800, 1000, 1200, 1400, 7000]

# Plot frequency
frequency = data['total_bedrooms'].value_counts(bins=categories, sort=False)
freq = []
for num in frequency:
    freq.append(int(num))
df['frequency'] = freq
# df = df.astype(dtype='int64', errors='ignore')
plt.bar(df.index.to_list(), df['frequency'])
plt.title("Frequency of each category")
plt.show()

# Plot mean
mean = []
for i in range(len(categories)-1):
    arr = []
    j = 0
    for j in range(len(data_total_bedrooms)):
        num = data_total_bedrooms.iloc[j]
        if (num >= categories[i]):
            if (num < categories[i+1]):
                arr.append(num)
            elif i == len(categories):
                arr.append(num)
    mean.append(round(sum(arr)/len(arr),2))
df['mean'] = mean
plt.bar(df.index.to_list(), df['mean'])
plt.title("Mean of each category")
plt.show()
```

```python
# 6. Class
import turtle as t

class Flower:
    size = 10
    color = 'black'
    flower_leaves = 4
    
    def blossom(self, water):
        self.size = self.size + water
        self.color = self.color
        self.flower_leaves = self.flower_leaves
        t.shape('turtle')
        t.color(self.color)
        for i in range(self.flower_leaves):
            t.circle(self.size)
            t.right(360/self.flower_leaves)
            
    def flourish(self, water, sunlight):
        self.size = water * self.size
        self.flower_leaves = sunlight + self.flower_leaves
        while sunlight:
            color = ['black', 'red', 'yellow', 'green', 'purple', 'orange']
            self.color = color[sunlight]
            t.shape('turtle')
            t.color(self.color)
            for i in range(self.flower_leaves):
                t.circle(self.size)
                t.right(360/self.flower_leaves)
            sunlight -= 1
            self.size -= 2
            self.flower_leaves -= 2
            
    def fade(self):
        t.clear()

flower1 = Flower()
flower1.blossom(2)
flower1.flourish(5, 4)
flower1.fade()
```
